```python
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Import the given “Salary_Data.csv”
salariesData = pd.read_csv('Salary_Data.csv')
```

```python
# split the data into training and testing
X = salariesData.iloc[:, :-1].values 
Y= salariesData.iloc[:, 1].values
```

```python
# splitting 1/3 of the data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)# Fitting Simple Linear Regression to the training set

reg = LinearRegression()
reg.fit(X_train, Y_train)
```

```python
# Predicting the Test set result 
pred = reg.predict(X_test)
```

```python
# Calculating the Mean_squared_error
mse = mean_squared_error(Y_test, pred)
```

```python
# Visualising the training set results and test set results
plt.scatter(X_train, Y_train, color = 'blue')
plt.scatter(X_test, Y_test, color = 'red')

plt.title('Salary Data')

plt.xlabel('Experience  (Years)')
plt.ylabel('Salary')

plt.show()
```
